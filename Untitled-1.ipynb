```python
from   kmer_ml.logger import msg

from   scipy import sparse as sp

import numpy  as np
import pandas as pd
import gzip
import json
import os

class DataLoader():

    def __init__(self, path):
        self.path = path
        self.dirs = sorted(folder for folder in os.listdir(path) if os.path.isdir(os.path.join(path, folder)))

    def load_X(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            paths = [os.path.join(self.path, str(sub)) for sub in subs]
            return load_X(paths, sparse)
        return load_X(self.path, sparse)

    def load_Y(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return load_Y(os.path.join(self.path, subs))
        return load_Y(self.path, sparse)

    def load_XY(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return (f(os.path.join(self.path, subs)) for f in (self.load_X, self.load_Y))
        return (f(self.path, sparse) for f in (self.load_X, self.load_Y))

    def __check_sub(self, subs):
        for sub in subs:
            if str(sub) not in self.dirs:
                raise Exception()

def _load_X(path, sparse):
    with open(os.path.join(path, 'meta.json')) as f:
        meta = json.load(f)
        n, m, v = meta.values()
    
    with gzip.open(os.path.join(path, 'X.txt.gz')) as gz:
        if sparse:
            r, c, d = (np.zeros(v, dtype = np.uint32) for _ in range(3))
            N       = 0
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                N_1    = N + len(idx)
                r[N:N_1], c[N:N_1], d[N:N_1] = idx, i, values[idx]
                N      = N_1
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return sp.csr_matrix((d, (r, c)), shape = (n, m))
        else:
            ret = np.zeros((n, m), dtype = np.uint32)
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                ret[idx,i] = values[idx]
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return ret
    
def load_X(paths, sparse):
    if len(paths) == 1:
        return _load_X(paths[0], sparse)
    vstack = sp.vstack if sparse else np.vstack
    return vstack([_load_X(path, sparse) for path in paths])


def _load_Y(path, target = None):
    df = pd.read_csv(os.path.join(path, 'Y.txt'))
    if target:
        return pd.get_dummies(df[target])
    return df

def load_Y(paths, target = None):
    dfs = [_load_Y(path, False) for path in paths]
    if target:
        dfs = [df[target] for df in dfs]
        return pd.get_dummies(pd.concat(dfs, axis = 0, ignore_index = True))
    return pd.concat(dfs, axis = 0, ignore_index = True)
```

```python
import gzip
import numpy as np
import json
import os
import gc

from   multiprocessing import Pool
from   biolearn.logger  import Counter, msg

import pandas as pd

path = 'data-by-year/2014'

def read_X(path, mask = None, dtype = np.uint8, n_processes = 8, verbose = True):
    if verbose: msg('opening resources', inline = True)
    global COUNTER, MASK, DTYPE, ARRAY, I
    COUNTER = Counter('allocating line {:,d}', display = verbose)
    MASK    = mask
    DTYPE   = dtype
    with open(os.path.join(path, 'meta.json')) as f, gzip.open(os.path.join(path, 'X.txt.gz')) as gz, Pool(n_processes) as pool:
        meta    = json.load(f)
        n, m    = meta['n'], meta['m']
        ARRAY   = np.zeros((n, m), dtype = dtype)
        I       = 0
        if mask:
            m = sum(mask) if np.array(mask).dtype == np.bool_ else len(mask)
        if verbose: msg(f'expected array dims: ({n:,d}, {m:,d})')
        pool.starmap(allocate, enumerate(gz))
        return ARRAY

def allocate(j, line):
    global COUNTER, MASK, DTYPE, I
    COUNTER.increment()
    if j in MASK:
        array    = np.array(line.decode().rstrip().split())
        ARRAY[I] = array[MASK] if MASK else array
        del array; gc.collect()
        I       ++ 1

        # if DTYPE != np.uint16:
        #     ret = ret.astype(DTYPE)
        # return ret[MASK] if MASK else ret

# TODO: reduce memory usage when performing parallel reading computations (load_X)

class DataLoader():

    def __init__(self, path, dtype = np.uint16, n_processes = 8, verbose = True):
        self.path        = path
        self.dtype       = dtype
        self.n_processes = n_processes
        self.verbose     = verbose

        if 'X.txt.gz' not in os.listdir(path):
            self.subpaths = sorted(sub for sub in os.listdir(path) if os.path.isdir(os.path.join(path, sub)))
        else:
            self.subpaths = None

    def get_params(self):
        return {key : self.__dict__.get(key) for key in ['path',' dtype', 'n_processes', 'verbose']}

    def set_params(self, params):
        valid = ['path',' dtype', 'n_processes', 'verbose']
        for key, value in params.items():
            if key in valid:
                self.__dict__[key] = value

    def load_X(self, *paths, mask = None):
        if len(paths) == 0:
            return read_X(self.path, mask)

            
        N = 0
        for path in paths:
            join = os.path.join(self.path, str(path))
            with open(os.path.join(join, 'meta.json')) as f:
                 meta = json.load(f)
            n, m = meta['n'], meta['m']
            if mask:
                m  = sum(mask) if np.array(mask).dtype == np.bool_ else len(mask)
                N += n
            
        ret = np.zeros((N, m), dtype = self.dtype)
        N   = 0
        for path in paths:
            join = os.path.join(self.path, str(path))
            with open(os.path.join(join, 'meta.json')) as f:
                 meta = json.load(f)
            n, m = meta['n'], meta['m']
            if self.verbose:
                msg(f'reading "{join}"', inline = True)
            ret[N:N + n] = read_X(join, mask, self.dtype, self.n_processes, False)
            if self.verbose:
                msg(f'read "{join}"')
            N += n
        return ret

    def load_Y(self, *paths, column):
        df = pd.concat([pd.read_csv(os.path.join(self.path, str(path), 'Y.txt')) for path in paths], ignore_index = True)
        Y  = pd.get_dummies(df[column])
        return Y
    
    def load(self, *paths, column, mask = None):
        return self.load_X(*paths, mask = mask), self.load_Y(*paths, column = column)

    @property
    def identifiers(self):
        pass

    @property
    def columns(self):
        with open(os.path.join(self.path, 'features.txt')) as f:
            return f.read().split()
            
dataloader = DataLoader('data')

# X = dataloader.load_X(*range(2014, 2020), mask = range(10000))
# read_X(path, list(range(1000000)), verbose = False, dtype = bool)
```
