```python
from   scipy import stats
import numpy as np

np.random.seed(0)

d = lambda x : np.random.randint(0, [[2, 2, 3, 2, 5], [2, 5, 2, 2, 2], [3, 3, 6, 5, 5]][x])

y = np.random.randint(0, 3, 10)

x = np.array(list(map(d, y)))

t = np.zeros(x.shape[1])
b = np.zeros_like(t)

mu_i = x.mean(axis = 0)

for j in set(y):
    mask  = y == j
    n_j   = mask.sum()
    mu_ij = np.mean(x[mask], axis = 0)
    s2_ij = np.var(x[mask], axis = 0)

    t    += n_j * (mu_ij - mu_i) ** 2
    b    += n_j * s2_ij

s = t / b

s.round(7)
```

```python
def clean(array):
    ret  = repr(array)
    neg1 = ['array(', ')', '[', ']']
    neg2 = [',', '       ']
    repl = {'\n' : r'\\' + '\n'}
    
    if len(array.shape) == 1:
        repl[' & '] = r'\\' + '\n'

    for neg in neg1:
        ret = ret.replace(neg, '')

    ret = ret.replace(', ', ' & ')

    for neg in neg2:
        ret = ret.replace(neg, '')

    for k, v in repl.items():
        ret = ret.replace(k, v)

    return ret

print(clean(s.round(7)))
```

```python
def _feature_selection(dataloader, init, inner_loop, outer_loop, values):
    ret, args, kwargs = init(dataloader)
    for value in values:
        for i, (x, label) in enumerate(dataloader.generator(value), 1):
            inner_loop(ret, i, x, label, value, *args, **kwargs)
        outer_loop(ret, value, *args, **kwargs)
    return ret

def fisher_score(dataloader, values):

    def init(dataloader):
        encode = {c : i for i, c in enumerate(set(dataloader.meta[dataloader.target]))}

        n = np.zeros(dataloader.c)
        m = np.zeros(dataloader.m)
        M = np.zeros((dataloader.c, dataloader.m))
        v = np.zeros((dataloader.c, dataloader.m))

        return {}, (encode, n, m, M, v), {}

    def inner_loop(ret, i, x, label, value, encode, n, m, M, v, **kwargs):
        print(f'\r{value} : {i}', end = '')
        y = encode[label]

        n[y] += 1
        m    += x
        M[y] += x
        v[y] += np.square(x)

    def outer_loop(ret, value, encode, n, m, M, v, **kwargs):

        print()

        n   = n.reshape(-1, 1)
        m   = m / dataloader.n
        M   = np.divide(M, n, where = n > 0)
        v   = np.divide(v, n, where = n > 0)
        v   = v - np.square(M)

        n   = n.reshape(-1)
        num = n @ np.square(M - m)
        den = n @ v

        S   = np.divide(num, den, where = (den > 0))

        ret[value] = S

    return _feature_selection(dataloader, init, inner_loop, outer_loop, values)

scores = fisher_score(dataloader, range(2014, 2020))
for key in list(scores):
    scores[f'2014-{key}'] = scores.pop(key)
```

```python
from scipy import stats

C      = []

def spearmanr_correlation(o1, o2):
    return stats.spearmanr(o1, o2).correlation

def spearmanr_pvalue(o1, o2):
    return stats.spearmanr(o1, o2).pvalue

def kendall_correlation(o1, o2):
    return stats.kendalltau(o1, o2).correlation

def kendall_pvalue(o1, o2):
    return 1 - stats.kendalltau(o1, o2).pvalue

def jaccard(o1, o2):
    common = len(set(o1) & set(o2))
    return common / (2 * len(o1) - common)

def common(o1, o2):
    return len(set(o1) & set(o2)) / len(o1)

for k in 10 ** np.arange(1, 6):

    matrix = np.zeros((6, 6))

    for i, o1 in enumerate(orders):
        for j, o2 in enumerate(orders[i:], i):
            matrix[i,j] = matrix[j,i] = common(o1[:k], o2[:k])

    C.append(matrix)

C = np.array(C)
```
