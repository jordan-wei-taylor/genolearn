```python
from   kmer_ml.logger import msg

from   scipy import sparse as sp

import numpy  as np
import pandas as pd
import gzip
import json
import os

class DataLoader():

    def __init__(self, path):
        self.path = path
        self.dirs = sorted(folder for folder in os.listdir(path) if os.path.isdir(os.path.join(path, folder)))

    def load_X(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            paths = [os.path.join(self.path, str(sub)) for sub in subs]
            return load_X(paths, sparse)
        return load_X(self.path, sparse)

    def load_Y(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return load_Y(os.path.join(self.path, subs))
        return load_Y(self.path, sparse)

    def load_XY(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return (f(os.path.join(self.path, subs)) for f in (self.load_X, self.load_Y))
        return (f(self.path, sparse) for f in (self.load_X, self.load_Y))

    def __check_sub(self, subs):
        for sub in subs:
            if str(sub) not in self.dirs:
                raise Exception()

def _load_X(path, sparse):
    with open(os.path.join(path, 'meta.json')) as f:
        meta = json.load(f)
        n, m, v = meta.values()
    
    with gzip.open(os.path.join(path, 'X.txt.gz')) as gz:
        if sparse:
            r, c, d = (np.zeros(v, dtype = np.uint32) for _ in range(3))
            N       = 0
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                N_1    = N + len(idx)
                r[N:N_1], c[N:N_1], d[N:N_1] = idx, i, values[idx]
                N      = N_1
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return sp.csr_matrix((d, (r, c)), shape = (n, m))
        else:
            ret = np.zeros((n, m), dtype = np.uint32)
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                ret[idx,i] = values[idx]
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return ret
    
def load_X(paths, sparse):
    if len(paths) == 1:
        return _load_X(paths[0], sparse)
    vstack = sp.vstack if sparse else np.vstack
    return vstack([_load_X(path, sparse) for path in paths])


def _load_Y(path, target = None):
    df = pd.read_csv(os.path.join(path, 'Y.txt'))
    if target:
        return pd.get_dummies(df[target])
    return df

def load_Y(paths, target = None):
    dfs = [_load_Y(path, False) for path in paths]
    if target:
        dfs = [df[target] for df in dfs]
        return pd.get_dummies(pd.concat(dfs, axis = 0, ignore_index = True))
    return pd.concat(dfs, axis = 0, ignore_index = True)
```

```python
def read_parallel(year, common = True, mask = None, dtype = np.uint16, indent = 0, processes = 64):
    global COUNTER, N, MASK, INDENT_
    MASK    = mask
    INDENT_ = indent
    fpath   = f'data/f{year}{"c" if common else ""}.txt'
    file    = fpath[5:]
    logger(f'reading "{file}"', indent = indent)
    COUNTER = Counter()
    with open(fpath) as f:
        if (year, common) in _cache:
            N, M    = _cache.get((year, common))
        else:
            line, N = count(f, indent = indent + 1)
            M       = len(line.split('\t'))
        m       = M if isinstance(mask, NoneType) else mask.sum() if mask.dtype.type == np.bool_ else len(mask)
        logger(f'dimensions : ({N:,d}, {M:,d}) -> ({N:,d}, {m:,d})', indent = indent + 1)
        with Pool(processes) as pool:
            ret = np.array(pool.map(allocate, f), dtype = dtype)
    logger(f'read "{file}"', 2, indent = indent)
    return ret

class Counter():

    def __init__(self):
        self.reset()
    
    def __repr__(self):
        return f"Counter(value = {self.value})"
    
    def increment(self, n = 1):
        with self.lock():
            self.val.value += n
        return self
    
    @property
    def value(self):
        return self.val.value
    
    def reset(self):
        value     = Value('i', 0)
        self.val  = value
        self.lock = value.get_lock
        return self

def allocate(line):
    global COUNTER, N, MASK, INDENT_
    update('allocating line', COUNTER.increment().value, N, indent = INDENT_ + 1)
    ret = np.array(line.split('\t'), dtype = np.uint16)
    return ret if isinstance(MASK, NoneType) else ret[MASK]
    
_cache = {(2014, False) : (145, 12103121),
          (2015, False) : (184, 12103121),
          (2016, False) : (929, 12103121),
          (2017, False) : (563, 12103121),
          (2018, False) : (638, 12103121),
          (2019, False) : (415, 12103121),
          (2014, True ) : (145,  9770777),
          (2015, True ) : (184,  9770777),
          (2016, True ) : (929,  9770777),
          (2017, True ) : (563,  9770777),
          (2018, True ) : (638,  9770777),
          (2019, True ) : (415,  9770777)}
```

```python
import gzip, os, numpy as np
from   kmer_ml.logger import msg


batch_size = 2 ** 11

i_file = 'data-sample/X.txt.gz'
o_file = 'test.txt.gz'

if o_file in os.listdir():
    os.remove(o_file)

with gzip.open(i_file) as gz_in, gzip.open(o_file, 'ab') as gz_out:
    N = gz_in.readline().decode().count(' ') + 1
    gz_in.seek(0)
    start = 0
    end   = batch_size
    batch = []
    while start < N:
        
        for line in gz_in:
            batch.append(line.decode().rstrip().split()[start:end])

        lines = '\n'.join(' '.join(row) for row in zip(*batch)) + '\n'

        gz_out.write(lines.encode())

        start = end
        end  += batch_size

        batch.clear()

        gz_in.seek(0)

        msg(start)
```

```python
import gzip
import numpy as np
import json
import os

from   multiprocessing import Pool
from   kmer_ml.logger  import Counter, msg

path = 'data-by-year/2014'

def read_X(path, mask = None, n_processes = 8, verbose = True):
    if verbose: msg('opening resources', inline = True)
    global COUNTER, MASK
    COUNTER = Counter('allocating line {:,d}', display = verbose)
    MASK    = mask
    with open(os.path.join(path, 'meta.json')) as f, gzip.open(os.path.join(path, 'X-t.txt.gz')) as gz, Pool(n_processes) as pool:
        meta    = json.load(f)
        n, m    = meta['n'], meta['m']
        if mask:
            m = sum(mask) if np.array(mask).dtype == np.bool_ else len(mask)
        if verbose: msg(f'expected array dims: ({n:,d}, {m:,d})')
        ret = np.array(pool.map(allocate, gz))
        return ret

def allocate(line):
    global COUNTER, MASK
    COUNTER.increment()
    ret = np.array(line.decode().rstrip().split(), dtype = np.uint16)
    return ret[MASK] if MASK else ret

read_X(path, list(range(100)), verbose = False)
```
