```python
from   kmer_ml.logger import msg

from   scipy import sparse as sp

import numpy  as np
import pandas as pd
import gzip
import json
import os

class DataLoader():

    def __init__(self, path):
        self.path = path
        self.dirs = sorted(folder for folder in os.listdir(path) if os.path.isdir(os.path.join(path, folder)))

    def load_X(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            paths = [os.path.join(self.path, str(sub)) for sub in subs]
            return load_X(paths, sparse)
        return load_X(self.path, sparse)

    def load_Y(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return load_Y(os.path.join(self.path, subs))
        return load_Y(self.path, sparse)

    def load_XY(self, *subs, sparse = False):
        if subs:
            self.__check_sub(subs)
            return (f(os.path.join(self.path, subs)) for f in (self.load_X, self.load_Y))
        return (f(self.path, sparse) for f in (self.load_X, self.load_Y))

    def __check_sub(self, subs):
        for sub in subs:
            if str(sub) not in self.dirs:
                raise Exception()

def _load_X(path, sparse):
    with open(os.path.join(path, 'meta.json')) as f:
        meta = json.load(f)
        n, m, v = meta.values()
    
    with gzip.open(os.path.join(path, 'X.txt.gz')) as gz:
        if sparse:
            r, c, d = (np.zeros(v, dtype = np.uint32) for _ in range(3))
            N       = 0
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                N_1    = N + len(idx)
                r[N:N_1], c[N:N_1], d[N:N_1] = idx, i, values[idx]
                N      = N_1
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return sp.csr_matrix((d, (r, c)), shape = (n, m))
        else:
            ret = np.zeros((n, m), dtype = np.uint32)
            for i, line in enumerate(gz):
                values = np.array(line.decode().strip('\n').split(), dtype = np.uint32)
                idx    = np.where(values)[0]
                ret[idx,i] = values[idx]
                msg(f'{i + 1:,d} of {m:,d}', inline = True)
            return ret
    
def load_X(paths, sparse):
    if len(paths) == 1:
        return _load_X(paths[0], sparse)
    vstack = sp.vstack if sparse else np.vstack
    return vstack([_load_X(path, sparse) for path in paths])


def _load_Y(path, target = None):
    df = pd.read_csv(os.path.join(path, 'Y.txt'))
    if target:
        return pd.get_dummies(df[target])
    return df

def load_Y(paths, target = None):
    dfs = [_load_Y(path, False) for path in paths]
    if target:
        dfs = [df[target] for df in dfs]
        return pd.get_dummies(pd.concat(dfs, axis = 0, ignore_index = True))
    return pd.concat(dfs, axis = 0, ignore_index = True)
```

```python
import scipy.sparse
from   multiprocessing import Pool
import numpy  as np
import pandas as pd
import os

class DataLoader():

    def __init__(self, path, meta_path = None, index = None):
        self.path  = path
        self.valid = [npz.replace('.npz', '') for npz in os.listdir(path) if npz.endswith('.npz')]

        if meta_path is None:
            self.meta = None
        else:
            self.load_meta(meta_path, index)
            self.meta.valid = list(set(self.valid) & set(self.meta.index))

        with open(os.path.join(path, 'features.txt')) as f:
            self.m = f.read().count(' ') + 1

    def load_meta(self, meta_path, index = None):
        df = pd.read_csv(meta_path)
        if index is not None:
            df = df.set_index(index)
        self.meta = df

    def _check_path(self, identifier):
        npz = os.path.join(self.path, identifier + '.npz')
        if os.path.exists(npz):
            return npz
        raise Exception(f'"{npz}" not a valid path!')

    def _check_meta(self, *identifiers, column = None):
        if self.meta is None:
            raise Exception('Meta data not loaded! Run the load_meta method first!')
        if column not in self.meta.columns:
            raise Exception(f'"{column}" not a valid column in self.meta!')
        if identifiers and self.meta.index.isin(identifiers).any():
            return self.meta.loc[identifiers, column]
        if identifiers and self.meta[column].isin(identifiers).any():
            return self.meta.loc[self.meta[column].isin(identifiers),column]
        return Exception()
        
    def _load_X(self, npz, features):
        npz = np.load(npz)
        data, col = npz['data'], npz['col']
        csr       = scipy.sparse.csr_matrix((data, (np.zeros_like(col), col)), shape = (1, self.m))
        return csr if features is None else csr[:,features]

    def get_identifiers(self, *values, column):
        self._check_meta(*values, column = column)            
        identifiers = self.meta.index[self.meta[column].isin(values)].values
        return identifiers

    def load_X(self, *identifiers, features = None, column = None):
        if column is None:
            npzs = (self._check_path(identifier) for identifier in identifiers)
            # with Pool(8) as pool:
            #     return sparse.vstack(pool.map(self._load_X, npzs))
            return scipy.sparse.vstack([self._load_X(npz, features) for npz in npzs])
        else:
            identifiers = self.get_identifiers(*identifiers, column = column)
            return self.load_X(*identifiers)

    def load_Y(self, *identifiers, column):
        return self._check_meta(*identifiers, column)

    def load(self, *identifiers, column):
        return self.load_X(*identifiers, column), self.load_Y(*identifiers, column)

def feature_selection(function, data_path, meta_path, target_column, index_column, group_column = None):
    dataloader = DataLoader(data_path, meta_path, index_column)
    npzs       = [npz.replace('.npz', '') for npz in os.listdir(data_path) if npz.endswith('.npz')]
    valid_ix   = dataloader.meta.index.isin(npzs)
    meta       = dataloader.meta.loc[valid_ix].reset_index()

    function(dataloader, meta, target_column, index_column, group_column)


def _fisher_score(dataloader, meta, target_column, index_column, group_column):
    if group_column is not None:
        ix_group_target = meta.loc[:,[index_column, group_column, target_column]].copy()
        scores = {group : {target : [0, 0, 0] for target in set(ix_group_target.loc[:,target_column])} for group in set(ix_group_target.loc[:,group_column])}
        for i in ix_group_target.index:
            ix, group, target = ix_group_target.loc[i]
            x = dataloader.load_X(ix)
            scores[group][target][0] += x
            scores[group][target][1] += x.power(2)
            scores[group][target][2] += 1
        return scores
    else:
        ix_target = meta.loc[:,[index_column, target_column]].copy()
        scores = {target : [0, 0, 0] for target in set(ix_target.loc[:,target_column])}
        for i in ix_target.index:
            ix, target = ix_target.loc[i]
            x = dataloader.load_X(ix)
            scores[target][0] += x
            scores[target][1] += x.power(2)
            scores[target][2] += 1
        return scores

def fisher_score_by_year(dataloader, meta, target_column, index_column, group_column):
    scores = _fisher_score(dataloader, meta, target_column, index_column, group_column)
    S      = {}

    for start in range(min(scores), max(scores)):
        years = range(start, max(scores))
        sub   = [0, 0]
        N     = {}
        for year in years:
            for label in scores[year]:
                if scores[year][label][2] == 0: continue
                sub[0] += scores[year][label][0].A
                sub[1] += scores[year][label][2]
                if label not in N:
                    N[label] = 0
                N[label] += scores[year][label][2]
        
        mean = sub[0] / sub[1]

        top  = 0
        bot  = 0

        for year in years:
            for label in scores[year]:
                if scores[year][label][2] == 0: continue
                mean_ij = scores[year][label][0].A / scores[year][label][2]
                mom2_ij = scores[year][label][1].A / scores[year][label][2]
                var_ij  = mom2_ij - mean_ij
                top += N[label] * np.square(mean_ij - mean)
                bot += N[label] * var_ij
        
        S[str(start)] = np.divide(top, bot, where = bot > 0)

        np.savez_compressed(os.path.join(dataloader.path, 'feature-selection', 'fisher-by-year.npz'), **S)
        
def localised_fisher_score_by_year(data_path, meta_path, target_column, index_column, group_column = None):
    dataloader = DataLoader(data_path, meta_path, index_column)
    npzs       = [npz.replace('.npz', '') for npz in os.listdir(data_path) if npz.endswith('.npz')]
    valid_ix   = dataloader.meta.index.isin(npzs)
    meta       = dataloader.meta.loc[valid_ix].reset_index()

    K          = [10, 100, 1000]
    R          = [10, 100, 1000, 10000, 100000]

    G          = {}
    end        = 2019
    for start in range(2014, 2019):
        years = list(range(start, end))
        met   = meta.loc[meta[group_column].isin(years)].set_index(index_column)

        sqdist = compute_sqdist(data_path, met)
        for k in K:
            G[k] = {}
            for r in R:
                G[k][r] = g = []
                g_bar   = list(range(dataloader.m))
                s_b     = np.zeros_like(dataloader.m)
                s_w     = np.zeros_like(dataloader.m)
                while len(g) < k:
                    J      = (s_b[g].sum() + s_b[g_bar]) / ((s_w[g] + 1e-8).sum() + s_w[g_bar])
                    k_star = np.argmax(J)
                    g.append(k_star)
                    g_bar.remove(k_star)



            
scores = feature_selection(fisher_score_by_year, 'testing', 'raw-data/meta-data.csv', 'Region', 'Accession', 'Year')


# dataloader = DataLoader('testing', 'raw-data/meta-data.csv', 'Accession')

# npzs = dataloader.get_identifiers(2018, column = 'Year')

# n      = len(npzs)
# sparse = np.zeros((n, n))

# batch  = 200

# m      = n // batch

# print(n, m)

# from scipy.spatial.distance import cdist

# for i in range(m):
#     start = i * batch
#     end   = start + batch
#     print(start, end)
#     Xi    = dataloader.load_X(*npzs[start:end], sparse = False)
#     sq    = cdist(Xi, Xi, metric = 'sqeuclidean')
#     ix    = np.arange(start, end)
#     sparse[ix,ix] = sq

#     for j in range(i + 1, m):
#         start_j = j * batch
#         end_j   = start_j + batch
#         print(start, end, start_j, end_j)
#         Xj      = dataloader.load_X(*npzs[start_j:end_j], sparse = False)
#         sq      = cdist(Xi, Xj, metric = 'sqeuclidean')
#         jx      = np.arange(start_j, end_j)
#         sparse[ix,jx] = sq

#     start_j += batch
#     Xj      = dataloader.load_X(*npzs[start_j:], sparse = False)
#     sq      = cdist(Xi, Xj, metric = 'sqeuclidean')
#     sparse[ix,start_j:] = sq

#     sq    = cdist(Xj, Xj, metric = 'sqeuclidean')
#     sparse[start_j:,start_j:] = sq
# X_train = dataloader.load_X(2014, 2015, 2016, 2017, 2018, column = 'Year', sparse = False)
# X_test  = dataloader.load_X(2019, column = 'Year')
# def compute_sqdist(dataloader):
```
