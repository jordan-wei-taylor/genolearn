```python
from genolearn import DataLoader
from genolearn.models.classification import RandomForestClassifier

import numpy as np
dataloader = DataLoader('data', 'raw-data/meta-data.csv', 'Accession', 'Region', 'Year')
```

```python
encode = {c : i for i, c in enumerate(set(dataloader.meta[dataloader.target]))}

n = np.zeros(dataloader.c)
m = np.zeros(dataloader.m)
M = np.zeros((dataloader.c, dataloader.m))
v = np.zeros((dataloader.c, dataloader.m))
```

```python
def score(dataloader, n, m, M, v):
    n   = n.reshape(-1, 1)
    m   = m / dataloader.n
    M   = np.divide(M, n, where = n > 0)
    v   = np.divide(v, n, where = n > 0)
    v   = v - np.square(M)

    n   = n.reshape(-1)
    num = n @ np.square(M - m)
    den = n @ v

    S   = np.divide(num, den, where = (den > 0))

    return S
    
scores = []
for year in range(2014, 2019):
    for i, (x, label) in enumerate(dataloader.generator(year)):
        print(f'\r{i}', end = '')
        y = encode[label]

        n[y] += 1
        m    += x
        M[y] += x
        v[y] += np.square(x)
    print()
    scores.append(score(dataloader, n, m, M, v))

scores = np.array(scores)
```

```python
from matplotlib import pyplot as plt

plt.hist(scores)
```
