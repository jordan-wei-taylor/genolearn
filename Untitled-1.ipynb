```python
from   genolearn.dataloader import DataLoader
from   genolearn.metrics    import intra_accuracy
from   matplotlib import pyplot as plt
import numpy as np

dataloader = DataLoader('data', 'raw-data/meta-data.csv', 'Accession', 'Region', 'Year')

def intra_accuracy(Y, Y_hat):
    ret    = {}
    unique = set(np.unique(Y)) | set(np.unique(Y_hat))
    for u in unique:
        ret[u] = (Y_hat[...,Y == u] == u).mean(axis = -1)
    return ret

accs       = {}
times      = {}
for year in range(2014, 2019):
    accs[year]  = {}
    times[year] = {}
    for model in ['lr', 'rf']:
        train, test        = dataloader.load_train_test_identifiers(range(year, 2019), [2019], 15)
        Y_train            = dataloader.load_Y(*train)
        Y_test             = dataloader.load_Y(*test)
        npz                = np.load(f'O157-region/{year}/{model}/results.npz')
        intra              = intra_accuracy(Y_test, npz['labels'][npz['predict']])
        train_counts       = dict(zip(*np.unique(Y_train, return_counts = True)))
        test_counts        = dict(zip(*np.unique(Y_test , return_counts = True)))
        p_train            = {key : val / len(train) for key, val in train_counts.items()}
        p_test             = {key : val / len(test ) for key, val in test_counts.items()}
        keys               = sorted(set(intra) | set(train_counts) | set(test_counts))
        accs[year][model]  = {key : [d.get(key, 0) for d in [intra, train_counts, test_counts, p_train, p_test]] for key in keys}
        times[year][model] = npz['time']

colors  = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
unique = sorted(set(sum([list(accs[year][model].keys()) for year in accs for model in accs[year]], [])))

fig, ax = plt.subplots(4, 2, figsize = (8, 14), sharey = True)

ax      = ax.flatten()

for i, u in enumerate(unique):
    for k in range(5):
        obj = accs[2014]['rf'][u]
        ax[i].plot(range(5, 105, 5), obj[0][k], c = colors[k], alpha = 0.4, ls = ':')
        ax[i].plot(range(5, 105, 5), obj[0][k].mean(axis = -1), c = colors[k], label = f'$10^{k + 2}$')
    ax[i].set_xticks([5] + list(range(20, 101, 20)))
    ax[i].set_xticklabels([5] + list(range(20, 101, 20)))
    ax[i].grid(ls = (0, (5, 5)))
    ax[i].annotate(f'{obj[1]} ({obj[3]:.1%})', (0.01, 1.03), xycoords = 'axes fraction', ha = 'left')
    ax[i].annotate(f'{obj[2]} ({obj[4]:.1%})', (0.99, 1.03), xycoords = 'axes fraction', ha = 'right')
    ax[i].annotate(u, (0.5, 1.03), xycoords = 'axes fraction', ha = 'center')
ax[-1].legend(ncol = 5, loc = (-0.9, -0.25))
```

```python
fig, ax = plt.subplots(8, 5, figsize = (12, 20), sharey = True, sharex = True)

for i, u in enumerate(unique):
    for j, year in enumerate(range(2014, 2019)):
        obj = accs[year]['rf']
        if u not in obj:
            ax[i,j].set_title('0 (0.0%)')
            continue
        
        ax[i,j].set_title(f'{obj[u][1]:,d} ({obj[u][3]:.1%})')

        for k in range(5):
            ax[i,j].plot(range(5, 101, 5), obj[u][0][k], c = colors[k], alpha = 0.2, ls = ':')
            ax[i,j].plot(range(5, 101, 5), obj[u][0][k].mean(axis = -1), c = colors[k])

    ax[i,0].set_ylabel(u)
    
    if u in obj:
        ax[i,-1].annotate(f'{obj[u][2]:,d} ({obj[u][4]:.1%})', (1.08, 0.5), xycoords = 'axes fraction', ha = 'center', va = 'center', rotation = 270)
```
